```python
# Turtle Game ursing the package 'turtle'
#Import relevant modules

import turtle
import random
import time

#Setting up a nice screen for the game
screen = turtle.Screen() #standard screen
screen.bgcolor('lightblue')  #usa schreibweise #Hintergrundfarbe hellblau


#we want two players in the game: whoever gets to the other side wins!

#Player one basic set up
player_one = turtle.Turtle()
# Color of player one
player_one.color('blue')
#make this player a turtle
player_one.shape('turtle')

#player two basic setup
player_two = player_one.clone()
#colour of player two
player_two.color('red')

#postion the players
player_one.penup()
player_one.goto(-300,200)
player_two.penup()
player_two.goto(-300,-200)

#draw the finish line
player_one.goto(300,-250)
player_one.left(90)
player_one.pendown()
player_one.color('black')
player_one.forward(500)
player_one.write('Finish!', font=24)

#player one goes back to starting postition
player_one.penup()
player_one.color('blue')
player_one.goto(-300,200)
player_one.right(90)

#make sure both players have the pen down so you can see who is moving
player_one.pendown()
player_two.pendown()

#create values for the die
die=[1,2,3,4,5,6]

# lets create the game!
for i in range(30):
    if player_one.pos() >= (300,250):
        print("Player one Winds the Race!")
        break
    elif player_two.pos() >= (300,-250):
        print("Player two Winds the Race!")
        break
    else:
        die_roll = random.choice(die)
        player_one.forward(30*die_roll)
        time.sleep(1) #dauert 1 sekunden
        die_roll2 = random.choice(die)
        player_two.forward(30*die_roll2)
        time.sleep(1)


turtle.done()  #Ã¶ffnet das turtle auf dem desktop wo man es sehen kann (muss als letztes stehen)
```
